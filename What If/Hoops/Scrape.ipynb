```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
import urllib
```

```python
def wis_scrape(team_id, view):
    user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
    
    url = f'https://www.whatifsports.com/nba-l/stats.asp?teamid={team_id}&view={view}'
    headers={'User-Agent':user_agent,}
    
    request=urllib.request.Request(url,None,headers) #The assembled request
    response = urllib.request.urlopen(request)
    data = response.read() # The data u need
    
    return pd.read_html(data)
```

# Input below

```python
team_id = 621520
league_id = 71608
```

```python
team = wis_scrape(team_id, 2)
team = team[0]
team.drop([12, 13], inplace = True)
team
```

```python
team['2PM'] = team['FGM'] - team['3PM']
team['2PA'] = team['FGA'] - team['3PA']
team['2P%'] = 100*(team['2PM']/team['2PA'])
team['3P%'] = 100*(team['3PM']/team['3PA']).fillna(0)
team['FG%'] = 100*(team['FGM']/team['FGA'])
team['eFG'] = (team['2PA']*team['2P%'] + 1.5*team['3PA']*team['3P%']) / team['FGA']
team.set_index('Name', inplace = True)
team
```

```python
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
    
url = f'https://www.whatifsports.com/nba-l/teamstats.asp?lid={league_id}&view=1&format=2&period=1'
headers={'User-Agent':user_agent,}
    
request=urllib.request.Request(url,None,headers) #The assembled request
response = urllib.request.urlopen(request)
data = response.read() # The data u need
    
team_stats = pd.read_html(data)
team_stats = team_stats[0]
team_stats.set_index('Team', inplace = True)
team_stats
```

```python
team_stats['2PM'] = team_stats['FGM'] - team_stats['3PM']
team_stats['2PA'] = team_stats['FGA'] - team_stats['3PA']
team_stats['efg'] = 100*(team_stats['2PM']  + 1.5*team_stats['3PM']) / team_stats['FGA']
team_stats.sort_values('efg', inplace = True, ascending = False)
team_stats
```
